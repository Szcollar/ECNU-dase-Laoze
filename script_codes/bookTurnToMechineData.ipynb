```python
import os
import re
import json
import threading
```

```python
def confirmDir(dir):
    if not os.path.exists(dir):
        os.mkdir(dir)
```

```python
def createDir(path):
    path = path.replace('\\','/')
    pathList = path.split('/')
    if pathList[-1] != '原文.txt':
        print(path)
        return
    dirPath = '.'
    for i in range(1,len(pathList)-1):
        dirPath = os.path.join(dirPath,pathList[i])
        confirmDir(dirPath)
```

```python
def modifyReadPath(path,string):
    path = path.replace('\\','/')
    if string == 'core':
        rpath = path.replace('books','afterClearCoreBooks')
    elif string == 'inCore':
        rpath = path.replace('首页','afterClear')
    return rpath
```

```python
def modifyWritePath(path,string):
    path = path.replace('\\','/')
    if string == 'core':
        wpath = path.replace('books','puncData\coreBooks')
    elif string == 'inCore':
        wpath = path.replace('ancient_chinese_books/首页','puncData\inCoreBooks')
    return wpath
```

```python
def modifyLossPath(path,string):
    path = path.replace('\\','/')
    if string == 'core':
        lpath = path.replace('books','puncData\coreBooksLoss')
    elif string == 'inCore':
        lpath = path.replace('ancient_chinese_books/首页','puncData\inCoreBooksLoss')
    return lpath
```

```python
inCoreHavePunIndexDir = './data/bookIndex/inCoreBook/havePun.txt'
coreHavePunIndexDir = './data/bookIndex/coreBook/havePun.txt'

inCoreNotHavePunIndexDir = './data/bookIndex/inCoreBook/notHavePun.txt'
coreNotHavePunIndexDir = './data/bookIndex/coreBook/notHavePun.txt'
```

```python
def getContent(path):
    with open(path,'r',encoding = 'utf-8') as txtFile:
        return txtFile.readlines()
```

```python
inCoreHavePunIndex = getContent(inCoreHavePunIndexDir)
coreHavePunIndex = getContent(coreHavePunIndexDir)

inCoreNotHavePunIndex = getContent(inCoreNotHavePunIndexDir)
coreNotHavePunIndex = getContent(coreNotHavePunIndexDir)
```

```python
temp = '又恐昏君#逃脱，预约定个、暗、号。若闻得响箭￥之声，便是昏君，即可追杀。这正是：'
findCharacter =re.compile(u"[\u4e00-\u9fa5]+")
findPunctuation = re.compile('[\u0020-\u007f\u3000-\u303f\u2000-\u206f\uff00-\uffef]')
# fundPun = re.compile()
subReduntPun = re.compile(u"[^\u4e00-\u9fa5。：；，？！]")
# findCharAndPun = re.compile()
print(re.findall(findCharacter,temp))
print(re.findall(findPunctuation,temp))
print(re.sub(subReduntPun,'',temp))
```

```python
def judgePun(paragraph):
    return len(re.findall(findPunctuation,paragraph)) > 0
```

```python
#tagSentence为修改处
```

```python
def tagSentence(sentence):
    tagList = []
    for i in range(len(sentence)-1):
        tagList.append(sentence[i]+'\t'+'0')
    tagList.append(sentence[-1]+'\t'+'1')
    return tagList
```

```python
def tagParagraph(paragraph):
    paragraph = re.sub(subReduntPun,'',paragraph)
    paraList = re.findall(findCharacter,paragraph)
    tagList = []
    for sentence in paraList:
        tagList.extend(tagSentence(sentence))
    return tagList
```

```python
def getDataFromOnePage(path):
    contentList = getContent(path)
    tagDict = {}
    withoutPunDict = {}
    for i in range(len(contentList)):
        if judgePun(contentList[i]):
            tagDict[i] = tagParagraph(contentList[i])
        else:
            withoutPunDict[i] = contentList[i]
    
    return tagDict,withoutPunDict
    
#     for key,value in tagDict.items():
#         print(key,'\t',value)
    
#     for key,value in withoutPunDict.items():
#         print(key,'\t',value)
#     print(len(tagDict.values())+len(withoutPunDict.values()))
#     print(tagDict)
#     print(WithoutPunDict)
```

```python
def writeLoss(path,dataDict,string):
    if string == 'core':
        with open('./data/puncData/coreLossParagraphLog.txt','a',encoding = 'utf-8') as txtFile:
            txtFile.write(path+'\n')
            for key in dataDict.keys():
                txtFile.write(str(key)+'\n')
            txtFile.write('\n')
    if string == 'inCore':
        with open('./data/puncData/inCoreLossParagraphLog.txt','a',encoding = 'utf-8') as txtFile:
            txtFile.write(path+'\n')
            for key in dataDict.keys():
                txtFile.write(str(key)+'\n')
            txtFile.write('\n')
    createDir(path)
    with open(path.replace('原文.txt','loss.json'),'w',encoding='utf-8') as jsonFile:
        json.dump(dataDict,jsonFile,indent=4,ensure_ascii=False)
```

```python
def writeTag(path,dataDict):
    createDir(path)
    with open(path,'w',encoding='utf-8') as txtFile:
        for key,valueList in dataDict.items():
            txtFile.write(str(key)+'\n')
            for value in valueList:
                txtFile.write(value+'\n')
```

```python
def handleCoreBooks():
    for path in coreHavePunIndex:
        print(path)
        path = path.rstrip('\n').replace('\\','/')    
        writePath = modifyWritePath(path,'core')
        lossPath = modifyLossPath(path,'core')
        readPath  = modifyReadPath(path,'core')
        tagDict,withoutPunDict = getDataFromOnePage(readPath)
        writeTag(writePath,tagDict)
        if len(withoutPunDict.keys()) == 0:
            continue
        writeLoss(lossPath,withoutPunDict,'core') 
```

```python
def handleIncoreBooks():
    for path in inCoreHavePunIndex:
        print(path)
        path = path.rstrip('\n').replace('\\','/')        
        readPath  = modifyReadPath(path,'inCore')
        if not os.path.exists(readPath):
            with open('./data/puncData/lossFile.txt','a',encoding='utf-8') as txtFile:
                txtFile.write(path+'\n')
            continue
        lossPath = modifyLossPath(path,'inCore')
        writePath = modifyWritePath(path,'inCore')
        tagDict,withoutPunDict = getDataFromOnePage(readPath)
        writeTag(writePath,tagDict)
        if len(withoutPunDict.keys()) == 0:
            continue
        writeLoss(lossPath,withoutPunDict,'inCore')
```
