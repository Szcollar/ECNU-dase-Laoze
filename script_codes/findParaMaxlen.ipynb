```python
import os
import re
```

```python
corePath = './data/afterClearCoreBooks/'
incorePath = './data/ancient_chinese_books/afterClear/'
```

```python
findCharacter =re.compile("[^\u4e00-\u9fa5]")
```

```python
def getParaslen(path):
    contents = []
    maxlen = 0
    with open(path,'r',encoding = 'utf-8') as txtFile:
        contents = txtFile.readlines()
    for para in contents:
        para = re.sub(findCharacter,'',para.rstrip('\n'))
        if maxlen < len(para):
            maxlen = len(para)
    return maxlen
```

```python
def getParaslenSet(path,Set):
    contents = []
    maxlen = 0
    with open(path,'r',encoding = 'utf-8') as txtFile:
        contents = txtFile.readlines()
    for para in contents:
        para = re.sub(findCharacter,'',para.rstrip('\n'))
        Set.add(len(para))
```

```python
def getParaslenDict(path,Dict):
    contents = []
    maxlen = 0
    with open(path,'r',encoding = 'utf-8') as txtFile:
        contents = txtFile.readlines()
    for para in contents:
        para = re.sub(findCharacter,'',para.rstrip('\n'))
        lenth = len(para)
        if lenth not in Dict:
            Dict[lenth] = 1
        else:
            Dict[lenth] += 1
```

```python
class node:
    def __init__(self,num):
        node.data = num
```

```python
lenthSet = set()
```

```python
def getMax(path,maxlen):
    if os.path.isfile(path):
        lenth = getParaslen(path)
        if maxlen.data<lenth:
            maxlen.data = lenth
        return
    pathList = os.listdir(path)
    for p in pathList:
        getMax(os.path.join(path,p),maxlen)
```

```python
def getMaxSet(path,Set):
    if os.path.isfile(path):
        getParaslenSet(path,Set)
        return
    pathList = os.listdir(path)
    for p in pathList:
        getMaxSet(os.path.join(path,p),Set)
```

```python
def getMaxDict(path,Dict):
    if os.path.isfile(path):
        getParaslenDict(path,Dict)
        return
    pathList = os.listdir(path)
    for p in pathList:
        getMaxDict(os.path.join(path,p),Dict)
```
