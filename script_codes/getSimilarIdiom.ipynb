```python
def splitIdioms(idioms):
    if '、' in idioms:
        return set(idioms.split('、'))
    elif ' ' in idioms:
        return set(idioms.split(' '))
    else:
        return set([idioms])
```

```python
def isRemove(idiom,dataDict,origion):
    if idiom not in dataDict:
        return False
    if origion in dataDict[idiom]:
        return True
    return False
```

```python
def getDictData(dataDict,csvData,index):
    for row in csvData:
        if len(row[index]) <= 1:
            continue
        idiomsSet = splitIdioms(row[index])
        temp = list()
        for idiom in idiomsSet:
            if isRemove(idiom,dataDict,row[0]):
                temp.append(idiom)
        for idiom in temp:
            idiomsSet.remove(idiom)
        if len(idiomsSet) == 0:
            continue
        if row[0] not in dataDict:
            dataDict[row[0]] = set()
        for idiom in idiomsSet:
            dataDict[row[0]].add(idiom)
```

```python
def writeInCsv(dataList,path):
    with open(path,'w',encoding='utf_8_sig',newline='') as csvFile:
        csvWriter = csv.writer(csvFile)
        for row in dataList:
            csvWriter.writerow(row)
```
