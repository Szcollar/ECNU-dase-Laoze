```python
import os
import re
```

```python
findPunctuation = re.compile('[\u0020-\u007f\u3000-\u303f\u2000-\u206f\uff00-\uffef]')
findCharacter =re.compile("[^\u4e00-\u9fa5]")
```

```python
incoreBookDir = './data/ancient_chinese_books/afterClear'
coreBookDir = './data/afterClearCoreBooks'
vocabPath = './data/puncData/vocab.txt'
```

```python
vocabDict = {}
```

```python
def getContent(path):
    with open(path,'r',encoding = 'utf-8') as txtFile:
        return txtFile.read()
```

```python
def addDict(content):
    content = re.sub(findCharacter,'',content)
    for char in content:
        if char not in vocabDict:
            vocabDict[char] = 0
        else:
            vocabDict[char] += 1
```

```python
def sortDir(path):
    if os.path.isfile(path):
        addDict(getContent(path))
        return
    fileList = os.listdir(path)
    for file in fileList:
        sortDir(os.path.join(path,file))
```

```python
vocabDict = {}
```

```python
sortDir(coreBookDir)
```

```python
sortDir(incoreBookDir)
```

```python
print(len(vocabDict.keys()))
```

```python
vocabList = [(key,value) for key,value in vocabDict.items()]
```

```python
vocabList.sort(key=lambda x: x[1],reverse = True) 
```

```python
for index,row in enumerate(vocabList):
    print(index,' ',row)
```

```python
with open(vocabPath,'w',encoding = 'utf-8') as txtFile:
    for row in vocabList:
        txtFile.write(row[0]+str(row[1])+'\n')
```
