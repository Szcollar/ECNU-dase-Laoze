```python
import os
```

```python
inCoreDir = './data/ancient_chinese_books/首页'
coreDir = './data/books'
```

```python
def isEqualLenInList(dateList):
    for i in range(1,len(dateList)):
        if len(dateList[i]) != len(dateList[i-1]):
            return False
    return True
```

```python
def haveLevel(path):
    if os.path.isfile(path):
        pathList = path.split('/')
        fatherPath = '.'
        for i in range(1,len(pathList)-2):
            fatherPath =fatherPath +'/'+ pathList[i]
        fileList = os.listdir(fatherPath)
        if isEqualLenInList(fileList):
            print(fileList)
            print(fatherPath)
            print('*'*100)
    else:
        fileList = os.listdir(path)
        for file in fileList:
            filePath = path + '/' + file
            haveLevel(filePath)
```

```python
haveLevel(inCoreDir)
```

```python
def getBookNum(path):
    if os.path.isfile(path):
        return 1
    else:
        sum = 0
        fileList = os.listdir(path)
        for file in fileList:
            if file == '译文.txt':
                continue
            filePath = os.path.join(path,file)
            sum += getBookNum(filePath)
        return sum
```

```python
class Index:
    def __init__(this,i = 0):
        this.num = i
    
    def getStrNum(this):
        ans = ''
        lenth = len(str(this.num))
        lenth = 5 - lenth
        for i in range(lenth):
            ans += '0'
        return ans + str(this.num)
```

```python
def setBookIndex(path,fatherPath,index,dataList,pre):
    if os.path.isfile(path):
        dataList.append((fatherPath.split('/')[-1],pre+index.getStrNum(),path))
        index.num += 1
    else:
        fileList = os.listdir(path)
        for file in fileList:
            if file == '译文.txt':
                continue
            filePath = path + '/' + file
            setBookIndex(filePath,path,index,dataList,pre)
```

```python
inCoreList = []
setBookIndex(inCoreDir,inCoreDir[0:-3],Index(),inCoreList,'11')
```
