```python
import os
import json
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent='myuseragent')
```

```python
location  = geolocator.geocode('樟树市, 临江镇, 宜春市, 樟树市, 江西省, 331200, 中国')
print(location.latitude,location.longitude)
```

```python
with open('./data/GISdata/GIS2.0.json','r',encoding = 'utf-8') as jsonFile:
    GISdata = json.load(jsonFile)
```

```python
print(GISdata)
```

```python
GIS3 = {}
for key,value in GISdata.items():
    if key in GIS3:
        continue
    location  = geolocator.geocode(value[0].split(',')[0])
    print(value[0])
    GIS3[key] = [value[0],location.latitude,location.longitude,int(value[1].rstrip('\n'))]
```

```python
print(GIS3)
```

```python
with open('./data/GISdata/GIS3.0.json','w',encoding = 'utf-8') as jsonFile:
    json.dump(GIS3,jsonFile,indent=4,ensure_ascii=False)
```
