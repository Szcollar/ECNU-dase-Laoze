```python
def getContent(path):
    with open(path,'r',encoding = 'utf-8') as txtFile:
        return txtFile.readlines()
```

```python
def removeEnter(contentList):
    i = 0
    while i<len(contentList):
        if contentList[i] == '\n' or len(contentList[i])==0:
            contentList.pop(i)
        else:
            i+=1
```

```python
def ratioOfUandChar(string):
    if len(string) == 0:
        return 1
    num = 0
    for char in string:
        if char == '\u3000':
            num +=1
    return num/len(string)
```

```python
def isRemove(string):
    string = string.rstrip('\n').strip(' ')
    ratio = ratioOfUandChar(string)
    if ratio > 0.35:
        return True
    if ratio < 0.08:
        return False
    string = string.strip('\u3000')
    stringList = string.split('\u3000')
    if len(stringList) <= 1:
        return True
    lenthList = []
    for row in stringList:
        lenthList.append(len(row))
    lenthSet = set(lenthList)
    if 0 in lenthSet:
        if len(lenthSet) == 2:
            return False
        else:
            return True
    if len(lenthSet) == 1:
        return False
    return True
    
#     if len(stringList)>1:
#         return True
#     return False
```

```python
def confirmDir(dir):
    if not os.path.exists(dir):
        os.mkdir(dir)
```

```python
def createDir(path):
    path = path.replace('\\','/')
    pathList = path.split('/')
    if pathList[-1] != '原文.txt':
        print(path)
        return
    dirPath = '.'
    for i in range(1,len(pathList)-1):
        dirPath = os.path.join(dirPath,pathList[i])
        confirmDir(dirPath)
```

```python
def writeInTxt(content,path):
    with open(path,'a',encoding = 'utf-8') as txtFile:
        txtFile.write(content)
        txtFile.write('\n')
```

```python
def modifyPath(path):
    path = path.replace('\\','/')
    path = path.replace('首页','afterClear')
    return path
```

```python
def modifyCorePath(path):
    path = path.replace('\\','/')
    path = path.replace('books','afterClearCoreBooks')
    return path
```

```python
corePathFileList = os.listdir(coreIndexDir)
for corePathFile in corePathFileList:
    pathList = getContent(os.path.join(coreIndexDir,corePathFile))
    print(len(pathList))
```

```python
inCorePathFileList = os.listdir(inCoreIndexDir)
for inCorePathFile in inCorePathFileList:
    pathList = getContent(os.path.join(inCoreIndexDir,inCorePathFile))
    for path in pathList:
        print(path)
        path = path.rstrip('\n')
        targetPath = modifyPath(path)
        createDir(targetPath)
        txtContentList = getContent(path)
        removeEnter(txtContentList)
        for row in txtContentList:
            row = row.rstrip('\n').strip(' ')
            if isRemove(row):
                continue
            writeInTxt(row.replace('\u3000',''),targetPath)
```
